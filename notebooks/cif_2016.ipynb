```python
from kats.consts import TimeSeriesData
import pandas as pd
from matplotlib import pyplot as plt
```

```python
filenames = [f"../datasets/external/cif_2016_t{i}.csv" for i in range(1, 9)]
raw_datasets = [pd.read_csv(filename) for filename in filenames]
```

```python
datasets_transposed = [pd.DataFrame(data.transpose().index, columns=[idx], dtype=float) for idx, data in enumerate(raw_datasets)]
data = pd.DataFrame(pd.concat(datasets_transposed, axis=1))
print(data.info())
print(data.describe())
print(data.head())
```

```python
for column in data.columns:
    print(column)
    data[column].plot()
    plt.show()
```
